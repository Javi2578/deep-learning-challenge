## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
df_dropped = application_df.drop(['EIN'], axis=1)
df_dropped.head()
```

```python
# Determine the number of unique values in each column.

unique_values = df_dropped.nunique()

# Display the number of unique values for each column
unique_values
```

```python
# Look at NAME value counts for binning
name_type_counts = df_dropped['NAME'].value_counts()
# Display the value counts for APPLICATION_TYPE
name_type_counts[name_type_counts >6]
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 6
name_types_to_replace = name_type_counts[name_type_counts <= cutoff_value].index.tolist()

# Replace in dataframe
for app in name_types_to_replace:
    df_dropped["NAME"] = df_dropped["NAME"].replace(app,"OTHER")

# Check to make sure binning was successful
df_dropped["NAME"].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = df_dropped['APPLICATION_TYPE'].value_counts()
# Display the value counts for APPLICATION_TYPE
application_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 500
application_types_to_replace = application_type_counts[application_type_counts < cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    df_dropped['APPLICATION_TYPE'] = df_dropped['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
df_dropped['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = df_dropped['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = 1000
classifications_to_replace = classification_counts[classification_counts < cutoff_value].index.tolist()
# Replace in dataframe
for cls in classifications_to_replace:
    df_dropped['CLASSIFICATION'] = df_dropped['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
df_dropped['CLASSIFICATION'].value_counts()
```

```python
# @title Default title text
# Convert categorical data to numeric with `pd.get_dummies`
df_dummies = pd.get_dummies(df_dropped)

# Display the first few rows of the dataframe with dummies to confirm the conversion
df_dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
# use get dummies
X = df_dummies.drop(["IS_SUCCESSFUL"], axis='columns').values
y = df_dummies["IS_SUCCESSFUL"].values


# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, input_dim=len(X_train_scaled[0]), activation="relu"),
    tf.keras.layers.Dense(80, activation="relu"),
     tf.keras.layers.Dense(40, activation="relu"),
    tf.keras.layers.Dense(1, activation ="sigmoid")
])

# Train the model
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.fit(X_train_scaled, y_train,
    epochs=100,  # Number of epochs to train for
    batch_size=256,  # Batch size for training
    validation_split=0.2,  # Fraction of training data to use for validation
    verbose=1  # Show training log
)
```

```python
# Evaluate the model using the test data
test_loss, test_mae = model.evaluate(X_test_scaled, y_test)
print("Test loss:", test_loss)
print("Test MAE:", test_mae)
```
